```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from models.RegressãoLinear.GradienteDescendenteEstocastico.model import StochasticGradientDescent
```

```python
df = pd.read_csv("/home/jorge/Documentos/IF/Reconhecimento de Padrões/pattern-recognition-RP/data/raw/artificial1d.csv", header=None)
```

Separando X e Y

```python
X, y = df.iloc[:, 0], df.iloc[:, 1]
```

```python
X.shape, y.shape
```

Instanciando a classe e treinamento com X e y do dataset

```python
gd = StochasticGradientDescent()
gd.fit(X, y)
```

Gerando 1000 valores de X entre -1 e 1

```python
x_generated = pd.Series(index=range(1000))
x_generated[:] = np.random.uniform(-1, 1, 1000)
x_generated
```

```python
y_predicts = gd.predict(x_generated)
```

```python

plt.scatter(X, y, label='Dados');
plt.plot(x_generated, y_predicts, color='red', label='Reta de predição');
plt.title("Reta de Regressão")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.plot(gd.mse_history, label='MSE por Época', color='blue')
plt.title('Evolução do Erro Quadrático Médio (MSE)')
plt.xlabel('Épocas')
plt.ylabel('MSE')
plt.legend()
plt.grid()
plt.show()
```
